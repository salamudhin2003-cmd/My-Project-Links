```python
# Identifying missing values
df.isnull().sum()
```

```python
# Replacing "Null","Nan","" with np.nan
df.replace(["NULL","null","nan","NaN",""],np.nan, inplace = True)
```

```python
df["Transport_Mode"] = df["Transport_Mode"].fillna("Unknown")
```

```python
df["Transport_Mode"].unique()
```

```python
df["Weight_kg"].unique()
```

```python
# Convert negatives to absolute
df["Weight_kg"] = df["Weight_kg"].abs()
```

```python
# Replace 0 with NaN
df["Weight_kg"] = df["Weight_kg"].replace(0, np.nan)
```

```python
# Fill NaN with median
median_weight = df["Weight_kg"].median()
df["Weight_kg"] = df["Weight_kg"].fillna(median_weight)
```

```python
df["Weight_kg"].unique()
```

```python
df["Distance_km"].unique()
```

```python
df["Distance_km"] = df["Distance_km"].abs()
```

```python
median_distance = df["Distance_km"].median()
df["Distance_km"] = df["Distance_km"].fillna(median_distance)
```

```python
df["Distance_km"].unique()
```

```python
df["Vendor"].unique()
```

```python
df["Vendor"] = df["Vendor"].fillna("Unknown")
```

```python
df["Vendor"].unique()
```

```python
df["Cost"].unique()
```

```python
df["Cost"] = df["Cost"].abs()

df["Cost"] = df["Cost"].replace(0, np.nan)

median_cost = df["Cost"].median()
df["Cost"] = df["Cost"].fillna(median_cost)
```

```python
df["Cost"].unique()
```

```python
df["Selling_Price"].unique()
```

```python
df["Selling_Price"] = df["Selling_Price"].replace(0, np.nan)


df["Selling_Price"] = df["Selling_Price"].abs()

median_sp = df["Selling_Price"].median()
df["Selling_Price"] = df["Selling_Price"].fillna(median_sp)
```

```python
df["Selling_Price"].unique()
```

```python
df["Delivery_Status"].unique()
```

```python
df["Delivery_Status"] = df["Delivery_Status"].fillna("Unknown")
```

```python
df["Delivery_Status"].unique()
```

```python
df["Damage_Flag"].unique()
```

```python
df["Damage_Flag"] = df["Damage_Flag"].map({
    0: "No Damage",
    1: "Damaged"
})
```

```python
df["Damage_Flag"] = df["Damage_Flag"].fillna("Unknown")
```

```python
df["Damage_Flag"].unique()
```

```python
df["Delivery_Time_Days"].unique()
```

```python
df.loc[df["Delivery_Time_Days"] <= 0, "Delivery_Time_Days"] = np.nan

# Fill NaN with median
median_days = df["Delivery_Time_Days"].median()
df["Delivery_Time_Days"] = df["Delivery_Time_Days"].fillna(median_days)
```

```python
df["Delivery_Time_Days"].unique()
```

```python
df["Return_Reason"].unique()
```

```python
df["Return_Reason"] = df["Return_Reason"].fillna("Not Returned")
```

```python
df["Return_Reason"].unique()
```

```python
df["Driver_Rating"].unique()
```

```python
df["Driver_Rating"] = df["Driver_Rating"].replace(0, np.nan)
```

```python
median_rating = df["Driver_Rating"].median()
df["Driver_Rating"] = df["Driver_Rating"].fillna(median_rating)
```

```python
df["Driver_Rating"].unique()
```

```python
df["Vehicle_Type"].unique()
```

```python
df["Vehicle_Type"] = df["Vehicle_Type"].fillna("Unknown")
```

```python
df["Vehicle_Type"].unique()
```

```python
df["Fuel_Consumption_Litres"].unique()
```

```python
df["Fuel_Consumption_Litres"] = df["Fuel_Consumption_Litres"].abs()
```

```python
median_fuel = df["Fuel_Consumption_Litres"].median()
df["Fuel_Consumption_Litres"] = df["Fuel_Consumption_Litres"].fillna(median_fuel)
```

```python
df["Fuel_Consumption_Litres"].unique()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df["Shipment_Date"] = pd.to_datetime(df["Shipment_Date"], errors="coerce")
df["Delivery_Date"] = pd.to_datetime(df["Delivery_Date"], errors="coerce")
```

```python
df["Delivery_Time_Calc"] = (df["Delivery_Date"] - df["Shipment_Date"]).dt.days
```

```python
df["Delivery_Time_Calc"].unique()
```

```python
df["Delivery_Time_Calc"] = np.where(df["Delivery_Time_Calc"] < 0,np.nan, df["Delivery_Time_Calc"])
```

```python
# Verified
(df["Delivery_Time_Calc"] < 0).sum()
```

```python
df["Delivery_Time_Calc"].fillna(
    df["Delivery_Time_Calc"].median(),
    inplace=True
)
```

```python
df["Delivery_Time_Calc"].unique()
```

```python
df.drop(columns=["Delivery_Time_Days"], inplace=True)
```
