```python
# Compute summary statistics for Weight, Cost, and Selling_Price
columns = ['Weight_kg', 'Cost', 'Selling_Price']
summary_stats = df[columns].describe()
summary_stats
```

```python
# Finding outlier limits using IQR for Weight.
Q1 = df['Weight_kg'].quantile(0.25)
Q3 = df['Weight_kg'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print(f"Lower limit: {lower_limit}")
print(f"Upper limit: {upper_limit}")

# Identify outliers
outliers = df[(df['Weight_kg'] < lower_limit) | (df['Weight_kg'] > upper_limit)]
print(outliers)
```

```python
# Expected number of delays per week.
from scipy.stats import poisson,norm,binom
df['Shipment_Date'] = pd.to_datetime(df['Shipment_Date'])
df['Week'] = df['Shipment_Date'].dt.isocalendar().week
delays = df[df['Delivery_Status'] == 'Delayed']
weekly_delays = delays.groupby('Week').size()
# Poisson mean (expected delays per week)
lambda_delays = weekly_delays.mean()
print(f"Expected number of delays per week (Î»): {lambda_delays:.2f}")

# Example: probability of exactly 3 delays in a week
k = 3
prob_3_delays = poisson.pmf(k, lambda_delays)
print(f"Probability of exactly {k} delays in a week: {prob_3_delays:.4f}")

```
