```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/1a) Cleaned_logistics.csv")
df
```

```python
df.columns
```

```python
# Daily Shipment Count
df['Shipment_Date'] = pd.to_datetime(df['Shipment_Date'])
daily_count = df.groupby('Shipment_Date').size()

plt.figure(figsize=(16, 5))
plt.plot(daily_count.index, daily_count.values, linewidth=1)
plt.xlabel("Shipment Date")
plt.ylabel("Number of Shipments")
plt.title("Daily Shipment Count")
plt.show()
```

```python
# Vendor vs Shipments
vendor_count = df.groupby('Vendor').size()
plt.figure(figsize=(16, 5))
colors = ['skyblue', 'orange', 'green', 'red', 'purple', 'gray']
plt.bar(vendor_count.index, vendor_count.values,color = colors)
plt.xlabel("Vendor")
plt.ylabel("Number of Shipments")
plt.title("Vendor vs Shipments")
plt.show()
```

```python
#  Weight distribution
plt.figure(figsize=(16, 5))
plt.hist(df['Weight_kg'], bins=20, color='skyblue',edgecolor='black')
plt.xlabel("Weight (kg)")
plt.ylabel("Frequency")
plt.title("Weight Distribution of Shipments")
plt.show()
```

```python
# Delivery Status %
plt.figure(figsize=(16, 5))
status_counts = df['Delivery_Status'].value_counts()
plt.pie(status_counts, labels =status_counts.index, autopct='%1.1f%%',startangle=90,colors=['green', 'lightcoral', 'gold','skyBlue','orange','Violet'])
plt.title('Delivery Status Distribution')
plt.axis('equal')
plt.legend()
plt.show()
```
