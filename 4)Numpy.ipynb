```python
import pandas as pd
import numpy as np
```

```python
# Profit margin calculated using vectorized NumPy operation
profit_margin = (df["Selling_Price"] - df["Cost"]) / df["Cost"] * 100
profit_margin
```

```python
# Identifying extreme values using percentiles
low = np.percentile(df["Weight_kg"], 1)
low
```

```python
high = np.percentile(df["Weight_kg"], 99)
high
```

```python
extreme_outliers = df[(df["Weight_kg"] < low) | (df["Weight_kg"] > high)]
extreme_outliers
```

```python
# Simulating delivery delays using Poisson distribution
realistic_deliveries = df[df['Delivery_Time_Calc'] <= 30]
avg_delivery_days = realistic_deliveries['Delivery_Time_Calc'].mean()
simulated_delays = np.random.poisson(lam=avg_delivery_days, size=1000)
print(f"Average delivery time: {avg_delivery_days:.2f} days")
print(f"Sample of 10 simulated delays: {simulated_delays[:10]}")
print(f"Mean of simulated delays: {simulated_delays.mean():.2f} days")
```

```python
# Probability that a shipment is damaged(Binomial)


# Calculate probability that a shipment is damaged
p_damage = (df["Damage_Flag"] == "Damaged").mean()

# Binomial simulation (1 = Damaged, 0 = Not Damaged)
damage_simulation = np.random.binomial(n=1, p=p_damage, size=1000)

damaged_count = damage_simulation.sum()
damage_percentage = damaged_count / len(damage_simulation) * 100

print(f"Probability of a shipment being damaged: {p_damage:.2%}")
print(f"Damaged shipments in binomial simulation: {damaged_count} out of 1000")
print(f"Simulated damage percentage: {damage_percentage:.2f}%")
```
